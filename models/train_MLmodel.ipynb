```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

```python
pistachepd = pd.read_csv('Data_final.csv')

print(pistachepd.head())
```

```python
# Select data
clusterdata = pistachepd.filter(
        ['Gender', 'CLMclient', 'SeedPrice', 'Prep_Plow', 'Plant', 'Labor', 'Weed', 'Harvest_Labor', 'MarmitesHarvested'],
axis=1)

clusterdata.rename(columns={'Gender':'gender',
                                'CLMclient': 'clm_client',
                                'SeedPrice': 'seed_price',
                                'Prep_Plow': 'prep_plow',
                                'Plant': 'plant',
                                'Labor': 'labor',
                                'Weed': 'weed',
                                'Harvest_Labor': 'harvest_labor',
                                'MarmitesHarvested': 'marmites_harvested'},
                                inplace=True
)
```

```python
#Agglomerative Cluster model

hc = AgglomerativeClustering(n_clusters=2,
                             linkage='ward')

#y_hc = hc.fit_predict(clusterdata)
y_hc = hc.fit(clusterdata)
"""
    705 # non-optimized default implementation; override when a better
    706 # method is possible for a given clustering algorithm
--> 707 self.fit(X, **kwargs)
"""
clusterlist = hc.labels_
print(f"list of clusters: {clusterlist}")
```

```python
# ==== add the cluster to the initial df ====
pistachepd['Cluster'] = clusterlist
```

```python
# === create a model basedon the clustering

from sklearn.neighbors import KNeighborsClassifier
import joblib

neigh = KNeighborsClassifier(n_neighbors=2)
model = neigh.fit(X=clusterdata,
                    y=clusterlist)
joblib.dump(model, 'ag_model.pkl')
```
